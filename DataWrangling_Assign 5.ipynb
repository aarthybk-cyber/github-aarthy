### Merging the three Superstore sheets into a single DataFrame

**#### Loading the 'Orders' Sheet (first 5000 rows)**
PART B

```python
# Load the 'Orders' sheet, taking only the first 5000 rows
orders = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Assignments/Sample_Superstore.xls', sheet_name='Orders', nrows=5000)

# Display the first 5 rows of the 'Orders' DataFrame
display(orders.head())
```

#### Loading the 'People' Sheet

```python
# Load the 'People' sheet
people = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Assignments/Sample_Superstore.xls', sheet_name='People')

# Display the first 5 rows of the 'People' DataFrame
display(people.head())
```

#### Loading the 'Returns' Sheet

```python
# Load the 'Returns' sheet
returns = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Assignments/Sample_Superstore.xls', sheet_name='Returns')

# Display the first 5 rows of the 'Returns' DataFrame
display(returns.head())
```

```python
# Merge 'orders' with 'returns' using a left join on 'Order ID'
# We use a left join here to ensure that all orders are retained, and return information is added only for the orders that have been returned.
# For orders that were not returned, the 'Returned' column will contain NaN, which can be handled later if needed.
final_dataset = orders.merge(returns, on='Order ID', how='left')
# Storing merged final dataset to excel file
final_dataset.to_excel("/content/drive/MyDrive/Colab Notebooks/Assignments/Merged_Superstore.xlsx", index=False)

# Fill NaN values in 'Returned' column with 'No' (assuming orders not in 'returns' were not returned)
final_dataset['Returned'] = final_dataset['Returned'].fillna('No')

# Merge the result with 'people' using a left join on 'Region'
# A left join is appropriate here to keep all orders and their associated return status, and then append the regional manager's name.
# This ensures that all order details remain, and manager information is added where a matching region exists.
final_dataset = final_dataset.merge(people, on='Region', how='left')

print("Shape of the final dataset:", final_dataset.shape)
print("\nFirst 5 rows of the final dataset:")
display(final_dataset.head())

print("\nInformation about the final dataset:")
final_dataset.info()
```

### Managerial Request: Records from San Francisco with Sales Exceeding 1,000

```python
# Filter records where City is 'San Francisco' and Sales exceed 1,000
san_francisco_high_sales = final_dataset[(final_dataset['City'] == 'San Francisco') & (final_dataset['Sales'] > 1000)]

print(f"Number of records from San Francisco with sales exceeding 1,000: {len(san_francisco_high_sales)}")
print("\nQuick summary of these records (first 5 rows):")
display(san_francisco_high_sales.head())
```

```python
num_records = len(san_francisco_high_sales)
total_sales_value = san_francisco_high_sales['Sales'].sum()

print(f"Number of matching records: {num_records}")
print(f"Total Sales value for these records: ${total_sales_value:,.2f}")
```

### Insights from San Francisco High Sales Records

```python
# Analyze top categories and sub-categories for high sales in San Francisco
print("\nTop Categories in San Francisco High Sales:")
display(san_francisco_high_sales['Category'].value_counts())

print("\nTop Sub-Categories in San Francisco High Sales:")
display(san_francisco_high_sales['Sub-Category'].value_counts().head(5))

# Analyze profitability of these high sales
mean_profit = san_francisco_high_sales['Profit'].mean()
total_profit = san_francisco_high_sales['Profit'].sum()
print(f"\nAverage Profit per high-sales transaction in San Francisco: ${mean_profit:,.2f}")
print(f"Total Profit from high-sales transactions in San Francisco: ${total_profit:,.2f}")

# Analyze average sales and discounts
mean_sales = san_francisco_high_sales['Sales'].mean()
mean_discount = san_francisco_high_sales['Discount'].mean()
print(f"Average Sales amount per high-sales transaction in San Francisco: ${mean_sales:,.2f}")
print(f"Average Discount applied to high-sales transactions in San Francisco: {mean_discount:.2%}")
```
